## 1. Introduction

<a id='why_seaborn'></a>
### 1.1 Why Seaborn is Preferred in Data Science ? 

Seaborn is favored in data science for several reasons:
* Built-in Themes and Color Palettes; 
    * The package offers a variety of built-in themes and color palettes, allowing users to quickly customize the appearance of their visualizations to suit different contexts and preferences.

* Statistical Plotting; 
    * Seaborn simplifies the process of creating statistical plots by providing functions that directly operate on DataFrames. This streamlines the visualization process and makes it more accessible for data scientists.
* Integration with Pandas:
    * Seaborn seamlessly integrates with Pandas DataFrames, making it convenient to work with tabular data commonly used in data science tasks. This integration facilitates quick and efficient data exploration and analysis.
* Advanced Features:
    * The package includes advanced features for multi-plot grids, color mapping, and handling of categorical data. These features contribute to the versatility of Seaborn in addressing diverse data visualization requirements.

* Built on top of matplotlib

## 3. Distribution Plots for Continuous Variables

<a id='displot'></a>
### 3.1 Using the *displot* Function

```python
# Create a distribution plot for the not distracted column 
sns.displot(car_crash['not_distracted']);
```

```python
sns.displot(car_crash['not_distracted'], bins=10, fill = True);
```

> Another example for visualizing continuous variable distributions with the *displot* Function

```python
# Create a distribution plot for the total bill
sns.displot(tips['total_bill'], kde=True, color='skyblue')

# Set plot labels and title
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Total Bill Amount')

# Show the plot
plt.show()
```

> Here's an example where data is randomly generated for both columns and plotted on the same graph.

```python
data =np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
sns.displot(data, kde=True, color='skyblue', fill=True);
```

> Here's an example where two columns are plotted on two separate graphs.

```python
data =np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    sns.displot(data[col], kde=True, color='skyblue', fill=True);
```

<a id='jointplot'></a>
### 3.2 Exploring Relationships with the *jointplot* Function

* This function is designed to visualize the relationship between two variables. 
* It combines scatter plots for numeric variables and histograms for univariate distributions, providing a comprehensive view of the joint distribution of the data

```python
sns.jointplot(x='speeding',
              y='alcohol',
              data=car_crash);
```

```python
sns.jointplot(x='speeding',
              y='alcohol',
              data=car_crash,
              kind='kde');
```

> Another Example for created data 

```python

data =np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
with sns.axes_style('white'):
    sns.jointplot(x="x", y="y", data=data, kind='kde');
```

> Or Using 'hex' as kind of the plot

```python
data =np.random.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
with sns.axes_style('white'):
    sns.jointplot(x="x", y="y", data=data, kind='hex');
```

```python
sns.pairplot(car_crash);
```

## 4. Categorical Plots for categorical data

<a id='barplot'></a>
### 4.1 Visualizing Categorical Data with *barplot* function

```python
sns.barplot(x='sex',
            y='total_bill',
data=tips );
```

<a id='barplot_estimator'></a>
#### 4.1.1 The estimator for the *barplot*

* A Seaborn barplot visually represents the mean of the y parameter (or x if orient='h' is used) across a categorical or discrete numerical column. 
* The choice of statistical function can be modified using the estimator parameter, allowing the display of alternative statistics such as median, sum, minimum, maximum, etc."

```python
sns.barplot(x='sex',
            y='total_bill',
            data=tips,
            estimator='median'
            );
```

<a id='no_error_bars'></a>
#### 4.1.2 Omitting Error Bars

```python

sns.barplot(x='sex',
            y='total_bill',
            data=tips,
            estimator='sum',
            errorbar=None
            );
```

<a id='data_labels'></a>
#### 4.1.3 Including Data Labels

```python
ax = sns.barplot(x='sex',
                 y='total_bill',
                 data=tips,
                 estimator='sum',
                 errorbar=None,
                 );
ax.bar_label(ax.containers[0], fontsize=10);
```

<a id='clustering_barchart'></a>
#### 4.1.4 Creating a Clustered Bar Chart

```python
sns.barplot(x='sex',
            y='total_bill',
            data=tips,
            estimator='sum',
            errorbar=None,
            hue='smoker'
);
```

<a id='countplot'></a>
### 4.2 Visualizing Categorical Data with *countplot* function

* A Countplot is a straightforward graph used to tally the occurrences of a categorical variable in the dataset. 
* The hue parameter allows for a comparison between two categorical variables, making it ideal for observing how a variable influences the outcome in a classification problem.

```python
sns.countplot(x='sex', data=tips);
```

> Example to show the hue parameter effect 

```python
sns.countplot(x='sex', hue='smoker', data=tips);
```

> Another example : to show the hue parameter effect 

```python
# Create a barplot with the hue parameter
sns.barplot(x='day', y='total_bill', hue='sex', data=tips)

# Set plot labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill Amount')
plt.title('Total Bill Amount by Day with Gender Comparison')

# Show the plot
plt.show()
```

<a id='boxplot'></a>
### 4.3 Visualizing Categorical Data with *boxplot* function

* What boxplot can provides comparing to barplot 
    * Barplots, although less detailed compared to boxplots, offer simplicity in understanding, especially for the general public. 
    * In contrast, boxplots present multiple metrics on a single graph, including the median, quartiles, and interquartile range. 
    * While boxplots may be more challenging to comprehend than barplots, they provide a more intricate depiction of the data distribution, offering additional details.

```python
sns.boxplot(x='day',
            y='total_bill',
            hue='sex',
            data=tips
            );
```

* There are two functions similar to the boxplot are violinplot and boxenplot: 
    * A violinplot illustrates the distribution with a curve, akin to a kernel density estimation (KDE) plot. 
    * On the other hand, a boxenplot is a graph that displays more quantiles than the traditional boxplot.

<a id='violinplot'></a>
### 4.4 Visualizing Categorical Data with *violinplot* function

The violin plot is a combination of a box plot and a KDE plot.

```python
sns.violinplot(x='day',
               y='total_bill',
               hue='sex',
               data=tips
               );
```

<a id='boxenplot'></a>
### 4.5 Visualizing Categorical Data with *boxenplot* function
The boxenplot is a graph that displays more quantiles than the traditional boxplot.

```python
sns.boxenplot(x='day',
               y='total_bill',
               hue='sex',
               data=tips
               );
```

> The last two functions we'll explore are variations of the scatterplot.

* The scatterplot function may not yield optimal results with categorical data and discrete numeric data due to significant point overlap.

* To address this issue, we can turn to one of these two functions : 1) swarmplot and 2) stripplot.

* Swarmplot is more effective with a small number of points, while stripplot excels with a larger number of points. 
* The latter includes a parameter called "jitter" that enables additional spreading of the points for better visibility.

<a id='swarmplot'></a>
### 4.6 Visualizing Categorical Data with *swarmplot* function

```python
sns.swarmplot(x='day',
              y='total_bill',
data=tips );
```

```python
sns.swarmplot(x='day',
              y='total_bill',
              data=tips,
              color='red'
              );
```

<a id='stripplot'></a>
### 4.7 Visualizing Categorical Data with *stripplot* function

```python
sns.set_context('talk')
with sns.axes_style('dark'):
  sns.stripplot(x='day',
                y='total_bill',
                hue='sex',
                data=tips,
                jitter=0.2
                );
```

## 5. Matrix Plots Visualization 

* The matrix plot visualization is particularly useful for displaying relationships between variables in a matrix format. One common type of matrix plot is the heat map. 
* A heatmap is a graphical representation of data where individual values within a matrix are represented as colors. 
* It is an excellent tool for visualizing the correlation between different variables in a dataset.

```python
crash_corr = car_crash.corr(numeric_only=True)
crash_corr
```

<a id='heatmap'></a>
### 5.1 Heatmap for the Correlation Matrix

```python
sns.heatmap(crash_corr);
```

```python
sns.set_context('paper')
sns.heatmap(crash_corr);
```

<a id='heatmap_annotations'></a>
### 5.2 Heatmap with Annotations for the Correlation Matrix
> The `annot=True` parameter adds annotations to the heatmap, displaying the correlation values. 

```python
sns.heatmap(crash_corr, annot=True);
```

<a id='heatmap_palette_color'></a>
### 5.3 Controlling the Color Palette for the Heatmap

> The `cmap="coolwarm"` parameter sets the color palette

```python
sns.heatmap(crash_corr, annot=True, cmap='Blues');
```

<a id='heatmap_line_between_cells'></a>
### 5.4 Controlling the line between cells of the Heatmap
> The `linewidths=.5` parameter adds white lines between cells for better visibility.

```python
sns.heatmap(
    crash_corr,
    cmap='Blues',
    linecolor='white',
    linewidth=1
    );
```

## 6. Visualizing Data with Grids

* Seaborn provides different types of grids for organizing and visualizing data. Here are short notes about some of the commonly used grids along with examples:
* These grid functions in Seaborn offer flexibility and customization for organizing and visualizing data in various ways.
* We will see some of the commonly used grids in seaborn

<a id='Pairgrid'></a>
### 6.1 Pair Grid
* PairGrid is used for creating a grid of subplots for pairwise relationships in a dataset. 
* It's useful for exploring correlations and distributions.

```python
iris_grid = sns.PairGrid(iris, hue='species');
iris_grid.map(plt.scatter);
```

<a id='facetgrid'></a>
### 6.2 Facet Grid
* FacetGrid allows the creation of a grid of subplots based on the values of one or more categorical variables. 
* Each subplot displays a subset of the data.

```python
tips_fg = sns.FacetGrid(tips, col='time', row='smoker');
tips_fg.map(plt.hist, 'total_bill');
```

```python
tips_fg = sns.FacetGrid(tips, col='time', row='smoker');
tips_fg.map(plt.scatter, 'total_bill', 'tip');
```

```python
tips_fg = sns.FacetGrid(
    tips,
    col='time',
    hue='smoker',
    height=4,
    aspect=1.5
);
tips_fg.map(plt.scatter, 'total_bill', 'tip');
```

<a id='jointgrid'></a>
### 6.3 Joint Grid
* JointGrid is a specialized grid for joint plots, allowing the visualization of the joint distribution of two variables along with their individual distributions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset
tips_data = sns.load_dataset("tips")

# Create a JointGrid for the tips dataset
g = sns.JointGrid(data=tips_data, x="total_bill", y="tip")

# Plot a scatter plot on the grid
g.plot(sns.scatterplot, sns.histplot)

# Show the plot
plt.show()
```

<a id='regression_plots'></a>
## 7. Visualizing Linear Relationships (Regression Plots)

* For linear relationships we use `lmplot` function 
* *lmplot*  stands for "linear model plot" and is designed for visualizing the relationships between two numeric variables. 
* It provides a convenient interface to fit and visualize regression models.

* Important Additional Parameters for `lmplot` function:
    * hue: Allows for color-coding the data points based on a categorical variable.
    * col and row: Facet the plot by one or more categorical variables.
    * markers: Customize the markers used in the scatter plot.
    * order: Control the order of polynomial regression.

```python
sns.set_context('paper')
sns.lmplot(data=tips,
           x='total_bill',
           y='tip',
           hue='sex');
```

```python
sns.lmplot(data=tips,
           x='total_bill',
           y='tip',
           hue='sex',
           markers=['o', '*'],
           scatter_kws={'s':100, 'linewidths':0.5}
           );
```

```python
sns.lmplot(data=tips,
           x='total_bill',
           y='tip',
           col='sex',
           row='time');
```
